## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
fig, ax = plt.subplots(figsize=(10,6))
sns.swarmplot(x = "gender", y="age", data=kill ,hue="manner_of_death");
```

## 2: Plot 12 most dangerous cities

```python
y =kill.city.value_counts()[:12]
y
```

```python
# with seaborn barplot
plt.figure(figsize=(8,6))
sns.barplot(x=y.index, y=y.values , data=kill)
plt.xticks(rotation=45)
plt.title("Most Dangerous Cities", color ="b");
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(8,6))
sns.countplot(x = "city", data = kill, order = y.index)
ax.set_xticklabels(y.index, rotation =45)
ax.set_title("Most Dangereous City", color = "b", fontsize=14)
plt.show()
```

## 3: Plot 20 most dangerous states

```python
st = kill.state.value_counts()[:20]
st
```

```python
# with seaborn barplot
plt.figure(figsize=(8,6))
sns.barplot(x=st.index, y=st.values , data=kill)
plt.xticks(rotation=45)
plt.title("Most dangerous states", color ="b");
```

## 4: Plot Having mental ilness or not for killed people

```python
mental = kill["signs_of_mental_illness"].value_counts()
mental
```

```python
# with seaborn
sns.countplot("signs_of_mental_illness", data=kill);
```
