# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
---

```python
df.salary.value_counts()
```

```python
df.education.value_counts()
```

```python
higher_education = (df["education"] == "Bachelors" )| (df["education"] =="Masters") | (df["education"] =="Doctorate")
higher_education
```

```python
df_high_edu = df[(df["education"] == "Bachelors" )| (df["education"] =="Masters") | (df["education"] =="Doctorate")]
df_high_edu
```

```python
df_high_edu[df_high_edu["salary"] == ">50K"]
```

```python
count = df_high_edu["salary"][df_high_edu["salary"] == ">50K"].count()
```

```python
perc_high_edu
```

```python
print(f"Percentage of people with advanced education make more than 50K : {round(perc_high_edu * 100, 2)}")
```

 # What country has the highest percentage of people that earn >50K?
 ---


```python
country_salary_df = df[["native-country","salary"]][df["native-country"] != "?"].groupby("native-country").count()
country_salary_df
```

```python
country_salary_df_riches = df[["native-country","salary"]][(df["salary"] == ">50K") & (df["native-country"] != "?")].groupby("native-country").count()
country_salary_df_riches
```

```python
rich_per_country = (country_salary_df_riches / country_salary_df).sort_values(by="salary")
rich_per_country
```

```python
rich_per_country[rich_per_country["salary"] == rich_per_country["salary"].max()]
```

```python
#use iloc here
# rich_per_country.iloc[rich_per_country["salary"].argmax():(rich_per_country["salary"].argmax()+1),:]
rich_per_country.sort_values('salary',ascending=False).iloc[[0]]
```

```python
plt.figure(figsize=(14,7))
df.loc[df["salary"] == ">50K", "native-country"].value_counts().plot(kind="bar");
```

```python
plt.figure(figsize=(12,6))
df.loc[((df["native-country"] == "Iran")), "salary"].value_counts().plot.bar();
```
